```python
# import package to view the table format 
from tabulate import tabulate  
```

# Categories Table & Competitions Table Data Analysis queries ;

```python
# 1. List all competitions along with their category name

Mycursor.execute("select Competitions_table.*,Categories_table.category_name from SportRadar.Competitions_table left join SportRadar.Categories_table  on Competitions_table.category_id = Categories_table.category_id limit 10")
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(len(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql')))  
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 2. Count the number of competitions in each category

Mycursor.execute('select Categories_table.category_name ,count(Competitions_table.competition_id) as number_of_competations from SportRadar.Categories_table left join SportRadar.Competitions_table on Competitions_table.category_id = Categories_table.category_id group by Categories_table.category_name limit 10')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(len(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql')))  
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 3. Find all competitions of type 'doubles'

Mycursor.execute('select * from SportRadar.Competitions_table where type = "doubles" limit 10')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(len(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql')))  
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 4. Get competitions that belong to a specific category (e.g., ITF Men)

Mycursor.execute('select Competitions_table.*,Categories_table.category_name from SportRadar.Categories_table inner join SportRadar.Competitions_table on Competitions_table.category_id = Categories_table.category_id where Categories_table.category_name = "ITF Men" limit 10')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(len(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql')))  
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 5.Identify parent competitions and their sub-competitions
Mycursor.execute('''SELECT Parent.competition_id AS parent_competition_id,Parent.competition_name AS parent_competition_name,Sub.competition_id AS sub_competition_id,Sub.competition_name AS sub_competition_name FROM SportRadar.Competitions_table AS Parent LEFT JOIN SportRadar.Competitions_table AS Sub ON Parent.competition_id = Sub.parent_id WHERE Parent.parent_id IS NULL limit 10''')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(len(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql')))  
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 6.Analyze the distribution of competition types by category

Mycursor.execute('select Categories_table.category_name,Competitions_table.type,count(Competitions_table.type) as distribution from SportRadar.Categories_table join SportRadar.Competitions_table on Competitions_table.category_id = Categories_table.category_id group by Categories_table.category_name,Competitions_table.type limit 10')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(len(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql')))  
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 7. List all competitions with no parent (top-level competitions)

Mycursor.execute('select * from SportRadar.Competitions_table where Competitions_table.parent_id is null or Competitions_table.parent_id = ""')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(len(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql')))  
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

# Competitor_Rankings Table & Competitors Table Analysis Quaries

```python
# 1. Get all competitors with their rank and points.

Mycursor.execute('select competitors_table.*,Competitor_Rankings_table.rank,Competitor_Rankings_table.points from SportRadar.competitors_table inner join SportRadar.Competitor_Rankings_table on competitors_table.competitor_id = Competitor_Rankings_table.competitor_id limit 10')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(f"Number of rows fetched: {len(out)}")
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 2.Find competitors ranked in the top 5

Mycursor.execute('select competitors_table.name from SportRadar.Competitor_Rankings_table inner join SportRadar.competitors_table on Competitor_Rankings_table.competitor_id = competitors_table.competitor_id where Competitor_Rankings_table.rank <= 5 order by Competitor_Rankings_table.rank asc limit 5')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(f"Number of rows fetched: {len(out)}")
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 3.List competitors with no rank movement (stable rank)

Mycursor.execute('select competitors_table.name,Competitor_Rankings_table.movement from SportRadar.Competitor_Rankings_table inner join SportRadar.competitors_table on Competitor_Rankings_table.competitor_id = competitors_table.competitor_id where Competitor_Rankings_table.movement = 0 limit 10')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(f"Number of rows fetched: {len(out)}")
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 4.Get the total points of competitors from a specific country (e.g., Croatia)

Mycursor.execute('select competitors_table.name,competitors_table.country,sum(Competitor_Rankings_table.points) as total_points from SportRadar.competitors_table inner join SportRadar.Competitor_Rankings_table on competitors_table.competitor_id = Competitor_Rankings_table.competitor_id where competitors_table.country = "Croatia" GROUP BY competitors_table.country limit 10')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(f"Number of rows fetched: {len(out)}")
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 5.Count the number of competitors per country

Mycursor.execute('select country,count(competitor_id) as Total_competitors from SportRadar.competitors_table group by country limit 10')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(f"Number of rows fetched: {len(out)}")
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```

```python
# 6.Find competitors with the highest points in the current week

Mycursor.execute('select competitors_table.name,Competitor_Rankings_table.points as Highest_point from SportRadar.competitors_table inner join SportRadar.Competitor_Rankings_table on competitors_table.competitor_id = Competitor_Rankings_table.competitor_id where Competitor_Rankings_table.points=(select max(Competitor_Rankings_table.points) from SportRadar.competitors_table) limit 10')
# variable to store the fetched data
out=Mycursor.fetchall()   
#list comprasion to view the table format
print(f"Number of rows fetched: {len(out)}")
print(tabulate(out,headers=[i[0] for i in Mycursor.description],  tablefmt='psql'))  
```
