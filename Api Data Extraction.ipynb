Extracting data  as Datafrmae

1. Categories Table

```python
try:
    Categories_table=[]
    for Data in competations_data['competitions']:
        json_Categories={'category_id':Data.get('category',{}).get('id',None),'category_name':Data.get('category',{}).get('name',None)}
        Categories_table.append(json_Categories)
    Categories_table=pd.DataFrame(Categories_table)
    display(Categories_table)
except ValueError as json_err:
    print(f'json cant decode the file {json_err}')
finally:
    Categories_table.to_csv('Categories Table.csv',index=False)
    print('the has been saved')
```

2. Competitions Table

```python
try:
    Competitions_table=[]
    for element in competations_data['competitions']:
        json_Competitions={'competition_id':element.get('id',None),'competition_name':element.get('name',None),'parent_id':element.get('parent_id',None),'type':element.get('type',None),'gender':element.get('gender',None),'category_id':element.get('category',{}).get('id',None)}
        Competitions_table.append(json_Competitions)
    Competitions_table=pd.DataFrame(Competitions_table)
    display(Competitions_table)
except ValueError as json_err:
    print(f'json cant decode the file {json_err}')
finally:
    Competitions_table.to_csv('Competitions Table.csv',index=False)
    print('the file has been saved')
```

3. Complexes Table

```python
try:
    Complexes_table=[]
    for data in complexes_data.get('complexes',[]):
        json_data={'complex_id':data.get('id',None),'complex_name':data.get('name',None)}
        Complexes_table.append(json_data)
    Complexes_table=pd.DataFrame(Complexes_table)
    display(Complexes_table)
except ValueError as json_err:
    print(f'json cant decode the file {json_err}')
finally:
    Complexes_table.to_csv('Complexes Table.csv',index=False)
    print('the file has been saved')
```

4. Venues Table

```python
try:
    Venues_table=[]
    for data in complexes_data.get('complexes',[]):
        for list_data in data.get('venues',[]):
            json_data={'venue_id':list_data.get('id',None),'venue_name':list_data.get('name',None),'city_name':list_data.get('city_name',None),'country_name':list_data.get('country_name',None),'country_code':list_data.get('country_code',None),'timezone':list_data.get('timezone',None),'complex_id':data.get('id',None)}
            Venues_table.append(json_data)
    Venues_table=pd.DataFrame(Venues_table)
    display(Venues_table)
except ValueError as json_err:
    print(f'json cant decode the file {json_err}')
finally:
    Venues_table.to_csv('Venues Table.csv',index=False)
    print('the file has been saved')
```

5. Competitor_Rankings Table

```python
try:
    Competitor_Rankings_table=[]
    for data in Competitors_Rankings_data.get('rankings',[]):
        for list_data in data.get('competitor_rankings',[]):
            json_data={'rank_id':data.get('type_id',None),'rank':list_data.get('rank',None),'movement':list_data.get('movement',None),'points':list_data.get('points',None),'competitions_played':list_data.get('competitions_played',None),'competitor_id':list_data.get('competitor',None).get('id',None)}
            Competitor_Rankings_table.append(json_data)
    Competitor_Rankings_table=pd.DataFrame(Competitor_Rankings_table)
    display(Competitor_Rankings_table)
except ValueError as json_err:
    print(f'json cant decode the file {json_err}')
finally:
    Competitor_Rankings_table.to_csv('Competitor_Rankings Table.csv',index=False)
    print('the file has been saved')

```

6. Competitors Table

```python
try:
    competitors_table=[]
    for data in Competitors_Rankings_data.get('rankings',[]):
        for list_data in data.get('competitor_rankings',[]):
            json_data={'competitor_id':list_data.get('competitor',None).get('id',None),'name':list_data.get('competitor').get('name',None),'country':list_data.get('competitor',None).get('country',None),'country_code':list_data.get('competitor',None).get('country_code',None),'abbreviation':list_data.get('competitor',None).get('abbreviation',None)}
            competitors_table.append(json_data)
    competitors_table=pd.DataFrame(competitors_table)
    display(competitors_table)
except ValueError as json_err:
    print(f'json cant decode the file {json_err}')
finally:
    competitors_table.to_csv('Competitors Table.csv',index=False)
    print('the file has been saved')
```
